```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(url, sep = ',')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'Ano': brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['Ano'], dados_brasil['Imigrantes']);
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']);

#plt.yticks([500, 1000, 1500, 2000, 2500, 3000])

#Na aula é ensinado a retirar os textos acima do gráfico utilizando a ferramenta plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(dados_brasil['Ano'], dados_brasil['Imigrantes']);
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul.index, america_sul['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc = 'left', fontsize = 18)
ax.set_xlabel('Número de Imigrantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending = True)
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá entre 1980 e 2013', loc = 'left', fontsize = 18)
ax.set_xlabel('Número de Imigrantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette = palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize = 18)
  ax.set_xlabel('Número de Imigrantes', fontsize = 14)
  ax.set_ylabel('')

  plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.line(dados_brasil, x='Ano', y='Imigrantes', title='Imigração do Brasil para o Canadá entre 1980 e 2013')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(width=1000, height=500, xaxis={'tickangle':-45, 'showgrid':True, 'gridcolor':'black', 'mirror':True, 'linecolor':'black', 'linewidth':1},
                  yaxis={'showgrid':True, 'gridcolor':'black', 'mirror':True, 'linecolor':'black', 'linewidth':1}, xaxis_title='Ano', yaxis_title='Número de Imigrantes',
                  font_family='Arial', font_size=14, font_color='grey', title_font_color='black', title_font_size=22, plot_bgcolor='white')

#adicionei algumas customizações adicionais que encontrei na documentação, como 'showgrid':True, 'gridcolor':'black', 'mirror':True, 'linecolor':'black', 'linewidth':1, nos dicionários de xaxis e yaxis
#para adicionar grid, cor do grid e os 3 ultimos são para customizar a borda mais externa do gráfico, também alterei a cor do background com plot_bgcolor='white'.

fig.show()
```
